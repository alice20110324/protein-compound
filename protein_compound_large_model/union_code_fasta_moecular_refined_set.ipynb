```python
import pandas as pd 
import os
import csv
from Bio import PDB
import numpy as np
```

```python
pathes='dataset/affinity/refined-set/'

dirs=[]
for root, dirs, files in os.walk(pathes):
    #print(dirs)
    #dirs.append(dirs)
    dirs=dirs
    break
#print(dirs)
```

```python
#pdbfasta
#save as file namedd protein_sequences.fasta
#but the format is not right
#fasta
parser = PDB.PDBParser(QUIET=True)
ppb = PDB.PPBuilder()
save_fastas='dataset/affinity/refined-set/protein_sequences.fasta'
num=0
with open(save_fastas,'w+') as f_w:
    for i in dirs:
        num=num+1
        print('i in dirs:', i)
        subpath=os.path.join(pathes,i)
        #print(subpath)
        dirfiles=os.listdir(subpath)
        for  j in dirfiles:
            
            
            #print(j)
            if '_protein.pdb' in j:
                subfile=os.path.join(subpath,j)
                pdb_id=j[:4]
                f_w.write(pdb_id)
                f_w.write('\n')
                structure = parser.get_structure(pdb_id, subfile)  # PDB


                for pp in ppb.build_peptides(structure):
                    m=pp.get_sequence()
                    #print(pp.get_sequence())
                    f_w.write(str(m))
                    f_w.write('\n')
                    break
                   
                
 
```
