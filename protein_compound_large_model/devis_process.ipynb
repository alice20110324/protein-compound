```python
##change columns' name to the standard dataset. add 'code' name

import pandas as pd
affinity_davis=pd.read_csv('dataset/davis/davis_triplet.csv',sep=',')
print(affinity_davis.head())
rows,columns=affinity_davis.shape
print(rows)
```

```python
code_data=[i for i in range(1 , 30057)]
affinity_davis.drop(columns=['code'],inplace=True)
affinity_davis.insert(0, 'code', code_data, allow_duplicates = False)

print(affinity_davis.head())
```

```python
davis_old_columns=affinity_davis.columns
davis_new_columns=['code','smiles','sequence','logKa']
davis_dict_columns=dict(zip(davis_old_columns,davis_new_columns))
affinity_davis.rename(columns=davis_dict_columns, inplace=True)
print(affinity_davis.head())
```

```python
affinity_smiles_number=affinity_davis['smiles']
#print(davis_smiles_data)
ligands_smiles_number=ligands_can_new_dict.keys()


affinity_smiles_smiles=[ligands_can_new_dict[k] for k in affinity_smiles_number if k in ligands_smiles_number]
print(len(affinity_smiles_smiles))

affinity_davis['smiles']=affinity_smiles_smiles
print(affinity_davis.head())
```

```python
affinity_sequence_number=affinity_davis['sequence']
#print(davis_smiles_data)
protein_sequence_number=protein_new_dict.keys()


affinity_sequence_sequence=[protein_new_dict[k] for k in affinity_sequence_number if k in protein_sequence_number]
print(len(affinity_sequence_sequence))

affinity_davis['sequence']=affinity_sequence_sequence
print(affinity_davis.head())
```
